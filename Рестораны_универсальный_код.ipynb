```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
```

```python
headers = { 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 YaBrowser/24.1.0.0 Safari/537.36', 'Accept-Language': 'ru,en;q=0.9',  'Cache-Control': 'max-age=0', 'Sec-Fetch-Site': 'same-origin', 'Sec-Fetch-User': '?1',  'Sec-Fetch-Mode': 'navigate',  'Sec-Fetch-Dest': 'document', 'Upgrade-Insecure-Requests':
'1'}
```

```python
general_inf = []
for p in range(0, 600, 30):
  url_tr = f'https://www.tripadvisor.ru/FindRestaurants?geo=295424&offset={p}&establishmentTypes=10591&priceTypes=10955&broadened=true'
  for _ in range(15):
    res = requests.get(url_tr, headers = headers)
    if res.status_code == 200:
        break
    else :
      res = requests.get(url_tr, headers = headers)
  sleep(2)
  soup = BeautifulSoup(res.text, 'html')
  main_page =  soup.find_all('div', class_ = 'biGQs _P fiohW alXOW NwcxK GzNcM ytVPx UTQMg RnEEZ ngXxk')
  print(main_page)
  for t in main_page :
   rest_link = 'https://www.tripadvisor.ru' + t.find('a', class_ = 'BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS').get('href')
   if rest_link not in general_inf :
    general_inf.append(rest_link)
```

```python
len(general_inf)
```

```python
names_1 = []
price_1 = []
rate_1 = []

for i in general_inf:
  for _ in range(10):
    rest = requests.get(i, headers = headers)
    if rest.status_code == 200:
        break
    else:
      rest = requests.get(i, headers = headers)
  sleep(2)
  soup1= BeautifulSoup(rest.text, 'html')
  try:
    restic_name = soup1.find('h1', class_ = 'HjBfq').text
  except:
    restic_name = 'No name'
  print(restic_name)
  names_1.append(restic_name)
  try:
    rest_price_range = soup1.find('div', class_ = 'BMlpu').find_all('div', class_ = 'SrqKb')[0].text.replace("\xa0", "")
  except:
    rest_price_range = 'NaN'
  print(rest_price_range)
  price_1.append(rest_price_range)
  try:
    rating = soup1.find('div', class_ = 'QEQvp').find('span', class_ = 'ZDEqb').text.replace('\xa0', '')
  except:
    rating = 'NaN'
  print(rating)
  rate_1.append(rating)
```

```python
ht_1 = list (zip(names_1, price_1, rate_1))
rest_abu_dhabi = pd.DataFrame(ht_1, columns = ['names', 'price', 'rate'])
rest_abu_dhabi['price']  = rest_abu_dhabi['price'].apply(lambda x:  x.replace('руб', ''))
rest_abu_dhabi['rate'] =  rest_abu_dhabi['rate'].apply(lambda x: x.replace(',', '.'))
rest_abu_dhabi['type'] = 'Middle'
rest_abu_dhabi['rate'] = rest_abu_dhabi['rate'].apply(float)
rest_abu_dhabi['mean_rate'] = rest_abu_dhabi['rate'].mean()
```

```python
rest_abu_dhabi
```

```python
new_df = rest_abu_dhabi['price'].str.split('-' ,expand=True)
new_df = new_df.drop(3, axis =1)
new_df = new_df.drop(2, axis =1)
```

```python
new_df.columns = ['min_price', 'max_price']
new_df
```

```python
ch = 'а'
c = 'е'
h = 'и'
new_column_values = []
for i in new_df['min_price'].to_list():
  if ch in i:
    new_column_values.append(None)
  elif c in i:
    new_column_values.append(None)
  elif h in i:
    new_column_values.append(None)
  else:
    new_column_values.append(i)

new_df['min_price'] = new_column_values
```

```python
ch = 'а'
c = 'е'
h = 'и'
s ='н'
new_column_values = []
for i in new_df['max_price'].to_list():
  if i is not None:
    if ch in i:
      new_column_values.append(None)
    elif c in i:
       new_column_values.append(None)
    elif h in i:
       new_column_values.append(None)
    elif s in i:
       new_column_values.append(None)
    else:
       new_column_values.append(i)
  else:
     new_column_values.append(i)

new_df['max_price'] = new_column_values
```

```python
rest_abu_dhabi = pd.concat([rest_abu_dhabi, new_df], axis=1)
rest_abu_dhabi = rest_abu_dhabi.drop('price', axis = 1)
```

```python
min = []
for i in rest_abu_dhabi['min_price'] :
  if i == 'NaN':
    min.append(None)
  else:
    min.append(i)
rest_abu_dhabi['min_price'] = min
rest_abu_dhabi['min_price'] = rest_abu_dhabi['min_price'].astype('Int64')
```

```python
max = []
for i in rest_abu_dhabi['max_price'] :
  if i == 'NaN':
    max.append(None)
  else:
    max.append(i)
rest_abu_dhabi['max_price'] = max
rest_abu_dhabi['max_price'] = rest_abu_dhabi['max_price'].astype('Int64')
```

```python
rest_abu_dhabi['city'] = 'Abu Dhabi'
rest_abu_dhabi['country'] = 'UAE'
```
