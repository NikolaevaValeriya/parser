```python
all_hotels = []
all_prices = []
all_ratings = []
all_stars = []
all_lowest_prices = []
all_highest_prices = []
all_prices_range = []
city = []


for g in tqdm(cities_g):
    hotels = []
    oa_number = 0
    # if g == 293932:
    #     oa_number = 960
    # else:
    #     oa_number = 0
    oa_stop = None

    flag_stop = False

    while True:

        try:

            g_iter = str(cities_g[g])
            oa_iter = str(oa_number)
            url = 'https://www.tripadvisor.ru/Hotels-g' + g_iter + '-oa' + oa_iter

            for _ in range(15):
                response = requests.get(url, headers=headers)

                if response.status_code == 200:
                    break
                
            if response.status_code != 200:
                print(f'url {url} не нашелся ')


            soup = BeautifulSoup(response.text, 'html')
            name = soup.find_all('a', class_ = 'BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS')
            
            if oa_stop is None:
                oa_stop = int(soup.find('span', class_ = 'b').text.replace('\xa0', '').split(' ')[0])
            

            
            stop_count = 0

            for i in name:
                
                hotel = i.find('h3')
                if hotel != None:
                    hotel = hotel.text.split('. ')
                    

                    if hotel[1] not in hotels:
                        hotels.append(hotel[1])
                        all_hotels.append(hotel[1])

                        print(hotel[1])

                        url2 = 'https://www.tripadvisor.ru' + str(i['href'])
                        for _ in range(15):
                            response2 = requests.get(url2, headers=headers)
                            if response2.status_code == 200:
                                break
                        soup2 = BeautifulSoup(response2.text, 'html')

                        price = soup2.find('div', class_ = 'biGQs _P fiohW uuBRH')
                        if price != None:
                            all_prices.append(price.text)
                            # print(price)
                        else:
                            all_prices.append(np.nan)

                        rating = soup2.find('span', class_ = 'kJyXc P')
                        if rating != None:
                            all_ratings.append(rating)
                            # print(rating)
                        else:
                            all_ratings.append(np.nan)

                        s = soup2.find('svg', class_ = 'y d H0')
                        if s != None:
                            stars = s.find('title').text.split(' ')[0]
                            all_stars.append(stars)
                            # print(stars)
                        else:
                            all_stars.append(np.nan)

                        price_range = soup2.find('div', class_ = 'IhqAp Ci')
                        if price_range != None:
                            price_range = price_range.text.split(' (')[0]
                            lowest_price = price_range.split(' - ')[0]
                            highest_price = price_range.split(' - ')[1]
                            all_prices_range.append(price_range)
                            all_lowest_prices.append(lowest_price)
                            all_highest_prices.append(highest_price)
                            # print(price_range)
                            # print(lowest_price)
                            # print(highest_price)
                        else:
                            all_prices_range.append(np.nan)
                            all_lowest_prices.append(np.nan)
                            all_highest_prices.append(np.nan)

                        city.append(cities_g)

                        hotels.append(hotel[1]) 
                        print(hotel)

                    

                    
                    else:
                        stop_count += 1 
                        print(f'{stop_count} повторов отелей на странице')

                        if stop_count > 8:
                            flag_stop = True
                            print(f'flag_stop = True')
                            break

                        continue

                else:
                    print(f'Hotel is None, url = {url}')
                    continue
            
            if flag_stop:
                print('While остановился из-за повторов')
                break

            if oa_number > oa_stop:
                print('While остановился из-за oa_number > oa_stop')
                break

            oa_number += 30        


        except Exception as e:
            if flag_stop:
                break
            print(f' Исключение {e}')
            oa_number += 30
            continue
    
        
```
