```python
import pandas as pd
```

```python
df = pd.read_csv('coordinates.csv')
```

```python
df = df.rename(columns={'insurance, руб.': 'insurance', 'ticket_price, руб.': 'ticket_price'})
```

```python
df.head()
```

```python
df[['luggage_cost', 'flight_time_minutes', 'rest_price', 'lowest_hotel_price', 'highest_hotel_price']] = df[['luggage_cost', 'flight_time_minutes', 'rest_price', 'lowest_hotel_price', 'highest_hotel_price']].astype(int)
df['rest_rating'] = df['rest_rating'].round(2)
```

```python
df.to_csv('coordinates.csv', index=False)
```
