```python
# Импорт библиотек для работы с поисковыми запросами и сбором ссылок с сайта
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
```

# Написание пробного варианта скрипта, получение первых результатов

```python
# Используем метод get для получения кода страницы
url = requests.get('https://www.businessinsider.com/personal-finance/average-weekly-travel-insurance-rates')
```

```python
# Получаем содержимое запроса в байтах через метод content и выписываем код страницы, используя библиотеку BeautifulSoup
page = BeautifulSoup(url.content, "lxml")
print(page)
```

```python
# Используем метод find для поиска по странице (два параметра: раздел кода и класс)
insurance_cost = page.find('div', class_='horizontal-scroll')
# Используем метод text для преобразования отрывка кода страницы в текстовую форму
insurance_cost_text = insurance_cost.text
print(insurance_cost_text)
# Скрипт написан и готов к применению для различных сайтов и стран
```
